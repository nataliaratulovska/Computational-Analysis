```python
from gensim.models import KeyedVectors
#fasttextmodell deutsch unter https://fasttext.cc/docs/en/crawl-vectors.html
#https://radimrehurek.com/gensim/models/keyedvectors.html
fasttext = KeyedVectors.load_word2vec_format('cc.de.300.vec.gz', binary=False)
#binary kann weiter trainiert werden (zb auf eigenen datensatz)
#vec datei kann nicht weiter trainiert werden


wortliste = ['Universität', 'Zeugnis', 'Gebäude', 'Mensa', 'Kurs', 'Freunde', 'Artikel', 'Bibliothek', 'Kaffee', 'Mindestlohn']

for wort in wortliste:
    if wort in fasttext:
        n = fasttext.most_similar(wort, topn=5)
        print(f"\nNächste Nachbarn für '{wort}':")
        for nachbar, score in n:
            print(f"  {nachbar} ({score:.4f})")
    else:
        print(f"\n'{wort}' ist nicht im Vokabular des Modells.")
```
